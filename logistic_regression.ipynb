```python
# Hy my name is junaid ,lets start the project (ageInsurance model) from importing the necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

```

```python
# Lets load the dataset 
dataset=pd.read_csv(r"/home/junaid/Downloads/Book 3(Sheet1).csv")
dataset
```

```python
# Data  preprocessing (converting categorical data to numeric values using the replace() function)
dataset['bought_insurance'].replace({'No':'0','Yes':'1'},inplace = True)
dataset
```

```python
#(data visualization) lets visualize the dataset using scatterplot from matplotlib to visualize the relationship between two variables in the dataset
plt.scatter(x="age",y="bought_insurance",data=dataset)



plt.axhline(y=0.5,color='r',linestyle='--')
plt.xlabel("Age")
plt.ylabel("bought_inssurance")
plt.title("Age vs bought_inssurance")
plt.show()         # the datapoints above from the cutoff line these probablity is 1
                # and the datapoints below from cuttoff line these probability is 0
```

```python
# split the dataset into training and testing sets using the train_test_split function from scikit_learn
x_train,x_test,y_train,y_test=(train_test_split(dataset[['age']],dataset['bought_insurance'],test_size=0.2))
x_train
```

```python
# lets create the object of logistic regression and train the model using fit() method..
lr=LogisticRegression()

lr.fit(x_train,y_train)  # if the ouput becomes logisticregression that means our model is trained  successfully.
```

```python
# after model is tranied successfuly now our model is ready for predection
lr.predict([[60]])  # 1 means Yes
```
